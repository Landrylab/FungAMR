```python
#### Import packages needed (and some not needed....)

import pandas as pd
print(pd.__name__, pd.__version__)

import numpy as np
print(np.__name__, np.__version__)

import matplotlib as plt
print(plt.__name__, plt.__version__)

import scipy.stats as stats
import scipy
print(scipy.__name__, scipy.__version__)

import seaborn as sns
print(sns.__name__, sns.__version__)

import matplotlib.pyplot as plt
```

```python
#Open excel with entries converted with the MSA file, quite long to load

df_fungir = pd.read_excel("FungAMR_db_converted_pos.xlsx", header=0)
df_fungir
```

```python
##df excluding all confidence 3 score entries

df_not3 = df_fungir[(df_fungir["degree of evidence"] != 3) & (df_fungir["degree of evidence"] != -3)]
df_not3
```

## Cyp51 
### map on CaErg11 PDB 5V5Z

```python
df_erg11 = df_not3[df_not3["ortho_homolog"] == "Cyp51"]
df_erg11['resistant'] = df_erg11['degree of evidence'].apply(lambda x: 'yes' if x > 0 else 'no')
df_erg11 = df_erg11[np.isfinite(df_erg11['convert_pos'])]
df_erg11['convert_pos'] = df_erg11['convert_pos'].round(0).astype(int)
df_erg11
```

```python
# Grouping by 'resistant' and 'convert_pos', then counting occurrences
occurrences_df = df_erg11.groupby(['resistant', 'convert_pos']).size().reset_index(name='occurrences')

def find_min_value(x):
    return x.min()

# Group by 'resistant' and 'convert_pos', then apply the custom function to find the minimum value
best_df = df_erg11.groupby(['resistant', 'convert_pos'])['best degree of evidence'].agg(find_min_value).reset_index(name='best degree of evidence')
occurrences_df

best_df
```

```python
##Resistance (best confidence score)

pd.set_option('mode.chained_assignment', None) # Disable settingWithCopy warning

best_df = best_df[best_df["convert_pos"] >= 25]
best_df = best_df[best_df["convert_pos"] <= 524]
positions_to_remove = [430,431,432,433, 434,435, 436,437,438,439,440,441]

occ_df = best_df[~best_df["convert_pos"].isin(positions_to_remove)]
occ_df["convert_pos"] = ":" + occ_df["convert_pos"].astype(str)

occ_res_df = occ_df[occ_df["resistant"] == "yes"]


occ_res_df['empty_col'] = ''

df_chi = occ_res_df[['empty_col','convert_pos', 'best degree of evidence']]


            
custom_header = f'#\nattribute: score\nmatch mode: 1-to-1\nrecipient: residues\n'
with open(f'defattr_files/score_erg11_attr_final.txt', 'w') as fp:
    fp.write(custom_header)

df_chi.to_csv(f'defattr_files/score_erg11_attr_final.txt', sep='\t', header=None, index=None, mode='a')
 # Palette below is specified manually because converting seaborn palette to discrete colors excludes extreme values
 # Was done with online hexcolor picker so not 100% accurate, also ChimeraX still does not manage to re-create cmap as expected...
cmd = f'color byattribute score range 1, 8 palette #519497:#519497:#8361CB:#8361CB:#8361CB:#A67490:#A67490:#A67490'
print(cmd)
```
